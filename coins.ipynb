```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python

fraction_heads = 0.5
a = [fraction_heads]
steps = 4096

for i in range(steps):
  toss = random.randint(0, 1)
  n_trials = len(a) + 1 
  fraction_heads = ((fraction_heads*(n_trials - 1))/n_trials) + toss/n_trials
  a.append(fraction_heads ) 

plt.plot(a)
```

```python
num_trials = 100
num_coins = 30

a = np.random.rand(num_trials, num_coins)
b = np.array(a*2, dtype = int)
c = b.sum(axis=1)
plt.hist(c, bins=30)
plt.show()

```
