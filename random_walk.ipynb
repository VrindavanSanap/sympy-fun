```python
def get_walks(n_walks=1, n_steps = 100):
    """
    Returns simulations of n_walks walks with n_steps
    """

    directions = [-1,1]
    a = np.zeros((n_walks, n_steps))
    for i in range(n_walks):
        curr_loc = 0
        for j in range(n_steps):
            curr_loc += random.choice(directions) 
            a[i][j] = curr_loc
        return a
    

def simulate_walk(n_walks, n_steps):
    """ 
    simpulates walk for a given number of steps, then returns the distance from the startpoint
    """
    directions = [-1,1]
    distances = np.zeros(n_walks)
    for i in range(n_walks):
        curr_loc = 0 
        for _ in range(n_steps):
            curr_loc += random.choice(directions) 
        distances[i] = curr_loc
    return distances
```
